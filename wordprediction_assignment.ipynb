```python
#Installation of necessary libraries
import nltk
from nltk.corpus import brown
from nltk import ngrams
from collections import Counter
import pandas as pd

nltk.download('brown')
```

```python
# Cleaned tokenized text (remove punctuation)
tokenized_text = [word.lower() for word in brown.words() if word.isalpha()]

def generate_ngrams_df(tokens, n):
    n_grams = list(ngrams(tokens, n))
    freq_dist = Counter(n_grams)
    df = pd.DataFrame(freq_dist.items(), columns=['ngram', 'frequency']).sort_values(by='frequency', ascending=False)
    return df
```

```python
# Testing the function
generate_ngrams_df(tokenized_text, 2).head()
```

```python
# Carrying out a prediction
def predict_next_words(sequence, ngram_df, n=2, k=5):
    sequence = tuple(sequence.lower().split()[-(n-1):])  # last n-1 tokens
    filtered = ngram_df[ngram_df['ngram'].apply(lambda x: x[:n-1] == sequence)]
    top_k = filtered.sort_values(by='frequency', ascending=False).head(k)
    return top_k['ngram'].apply(lambda x: x[-1]).tolist()
```

```python
bigrams_df = generate_ngrams_df(tokenized_text, 2)
predict_next_words("the", bigrams_df, n=2, k=5)
```
